```python
category_feat_list=list()
for col in df_claim.columns:
    if df_claim[col].dtype=="object":
        category_feat_list.append(col)        
```

```python
from sklearn.preprocessing import LabelEncoder
for col in df_claim.columns:
    if df_claim[col].dtype=="object":
        df_claim[col]=LabelEncoder().fit_transform(df_claim[col])
```

```python
df_x=df_claim.drop(columns=["total_claim_amount"])
y=df_claim[["total_claim_amount"]]
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,df_x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV
lg = lgb.LGBMRegressor(silent=False)
param_dist = {"max_depth": [25,50, 75],
              "learning_rate" : [0.01,0.05,0.1],
              "num_leaves": [300,900,1200],
              "n_estimators": [200]
             }
grid_search = GridSearchCV(lg, n_jobs=-1, param_grid=param_dist, cv = 5, scoring="r2")
grid_search.fit(df_x,y)
grid_search.best_params_
```

```python
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
params={'learning_rate': 0.01,'max_depth': 25,'n_estimators': 200,'num_leaves': 300}
max_r_score=0
for r_state in range(42,100):
    x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = r_state,test_size=0.20)
    d_train = lgb.Dataset(x_train, label=y_train)
    lgb_model = lgb.train(params, d_train,categorical_feature=cate_feat_list)
    y_pred = lgb_model.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print("r2 score corresponding to ",r_state," is ",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
```

```python
from sklearn.model_selection import KFold
import numpy as np
scores=[]
k_fold = KFold(n_splits=5)
for train_index,test_index in k_fold.split(df_x):
    x_train, x_test, y_train, y_test = df_x.iloc[train_index], df_x.iloc[test_index], y.iloc[train_index], y.iloc[test_index]
    d_train = lgb.Dataset(x_train, label=y_train)
    lgb_model = lgb.train(params, d_train, categorical_feature = cate_feat_list)
    y_pred = lgb_model.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    scores.append(r2_scr)
print("Mean r2 score for lgbm regressor: ",np.array(scores).mean())
print("standard deviation in r2 score for lgbm regressor: ",np.array(scores).std())
print(scores)
```

```python
from sklearn.model_selection import cross_val_score
def model_evaluation(model,x,y):
    c_scores=cross_val_score(model,x,y,cv=5,scoring="r2")
    print("Mean r2 score for regressor: ",c_scores.mean())
    print("standard deviation in r2 score for regressor: ",c_scores.std())
    print("Cross Validation score",c_scores)
```

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
dtr=DecisionTreeRegressor()
r_state=maxr2_score(dtr,df_x,y)
```

```python
print("DECISION TREE\n\n")
model_evaluation(dtr,df_x,y)
```
